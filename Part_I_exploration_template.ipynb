### Clean up a couple of the data issues observed

```python
# Create a copy of gobike data set that cleaning will be performed on
df_clean = gobike.copy()
```

#### The data has some `datetime` columns as object. Clean this up by updating the datatype of `start_time` and `end_time` from object to datetime.

```python
df_clean['start_time'] = pd.to_datetime(df_clean['start_time'])
df_clean['end_time'] = pd.to_datetime(df_clean['end_time'])
```

#### Turn `user_type`, `member_gender`, `bike_share_for_all_trip` to categorical data type

```python
# convert user_type, member_gender into ordered categorical types
df_clean['user_type'] = df_clean['user_type'].astype('category')
df_clean['member_gender'] = df_clean['member_gender'].astype('category')
df_clean['bike_share_for_all_trip'] = df_clean['bike_share_for_all_trip'].astype('category')
```

```python
# Confirm that data type on the column has been converted to category.
df_clean.info()
```

#### So, from the ride `start_time`, extract new variables `start_year`, `start_month`, `start_month`, `start_day` and save them in their individual columns

```python
# convert the date to string using strftime and get the full month name using '%B'
df_clean['start_time_month_name'] = df_clean['start_time'].dt.strftime('%B')
```

```python
# get start time month number
df_clean['start_time_month']=df_clean['start_time'].dt.month.astype(int)
```

```python
# get start time week day
df_clean['start_time_week_day'] = df_clean['start_time'].dt.strftime('%A')
```

```python
# get start time day of the month
df_clean['start_time_day'] = df_clean['start_time'].dt.day.astype(int)
```

```python
# get start time hour of the day
df_clean['start_time_hour'] = df_clean['start_time'].dt.hour.astype(int)
```

#### Create a column for user's age based on the current year

```python
df_clean['age'] = 2019 - df_clean['member_birth_year']
```

#### Create a column for duration in minute

```python
df_clean['duration_min'] = df_clean['duration_sec']/60
```

## Univariate Exploration

#### 1. What is the distribution of Gender of the bike rider?

```python
# explore the gender feature of the gobike data
n_gender = df_clean['member_gender'].value_counts().sum()
base_color = sb.color_palette()[1]

gender_count = df_clean['member_gender'].value_counts()
gender_order = gender_count.index

sb.countplot(data = df_clean, x = 'member_gender', color = base_color, order = gender_order);
plt.xlabel('Gender Proportion');
plt.ylabel('Count');

# Logic to print the proportion text on the bars
for i in range (gender_count.shape[0]):
    # Remember, type_counts contains the frequency of unique values in the `type` column in decreasing order.
    count = gender_count[i]
    # Convert count into a percentage, and then into string
    gender_prop_string = '{:0.1f}%'.format(100*count/n_gender)
    # Print the string value on the bar. 
    plt.text(i, gender_count[i]+1200, gender_prop_string, va='center')
    
# specify the figure size
rcParams['figure.figsize'] = 8,5

 
```
